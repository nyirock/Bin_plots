```python
import os
def parse_bin(filename):
    handle = open(filename, )
    parsed = SeqIO.parse(handle, "fasta")


    #bin_name="".join(filename.split(".")[0]).title()
    bin_name = os.path.basename(filename)
    bin_name =  os.path.splitext(bin_name)[0]
    bin_name="".join(bin_name.split(".")[0:2])
    bin_name = bin_name[0].upper()+bin_name[1:]
    
    total = 0
    processed = 0
    names = list()

    for record in parsed:
        total += 1
        if total <10:
            print(record.id)

        name = record.id

        names.append(name)

    bin_info = pd.DataFrame({'name': names})
    bin_info['bin']=bin_name
    #     record_data = pd.DataFrame([{'name': name, 'length': length, 'coverage': coverage, 'gc': gc}])
    #     gc_data = pd.concat([gc_data, record_data], ignore_index=True)

    handle.close() 
    
    return bin_info
```

```python
def parse_contigs(filename):
    handle = open(filename)
    parsed = SeqIO.parse(handle, "fasta")


    column_names = ['name', 'length', 'coverage', 'gc']
    
    total = 0
    processed = 0
    names = list()
    lengths = list()
    coverages = list()
    gcs = list()
    for record in parsed:
        total += 1
        if total <10:
            print(record.id)
        #print(record.id), len(record.seq)
        #if operations[params["test"]](GC(record.seq), params["GC"]):
        #    processed += 1
        #    records.append(record)
        length = len(record.seq)
        gc = GC(record.seq)
        name = record.id
        name_parts = name.split("_")
        if len(name_parts) > 1: #we are looking for coverage in the end
            try: 
                coverage = name_parts[-1]
                coverage = float(coverage)
            except:
                print(coverage)
                coverage = None
        else:
            coverage = None

        lengths.append(length)
        names.append(name)
        coverages.append(coverage)
        gcs.append(gc)

    #     record_data = pd.DataFrame([{'name': name, 'length': length, 'coverage': coverage, 'gc': gc}])
    #     gc_data = pd.concat([gc_data, record_data], ignore_index=True)

    handle.close() 
    
    return pd.DataFrame({'name': names, 'length': lengths, 'coverage': coverages, 'gc': gcs})
```
